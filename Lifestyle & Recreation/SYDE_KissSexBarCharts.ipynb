```python
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['Had First Kiss','Had Sex']

NoBefore = [58, 81]
YesBefore = [33, 10]
NoAfter = [67, 76]
YesDone = [15, 8]
YesFirst = [7, 6]
total = [82, 84]


X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, NoBefore, 0.4, label = 'No')
plt.bar(X_axis - 0.2, YesBefore, 0.4, bottom = NoBefore, label = 'Yes')
plt.bar(X_axis + 0.2, NoAfter, 0.4, label = 'No')
plt.bar(X_axis + 0.2, YesDone, 0.4, bottom = NoAfter, label = 'Yes, done it before')
plt.bar(X_axis + 0.2, YesFirst, 0.4, bottom = total, label = 'Yes, for the first time')
  
plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of Students")
plt.title("Number of Students in each group")
plt.legend()
plt.show()
```
