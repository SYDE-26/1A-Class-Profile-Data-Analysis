```python
import matplotlib.pyplot as plt

Yes = [4,5,5,3,3,5,4,3,3,4,6,2,2,5,6,2,3,5,5,5,5,6,7,6,5,4,5,4,6,6,4,6,2,5,5,7,5,5,3,4,2,5,5,4,4,5,6,7,4,5,5,5,4,3,6,2,4,5,5,4,4,2,6,5,6,5,5,6,2,2,3,4,6,5,5,3,5,3,6]
No = [2,5,3,1,3,4,2,5,4,5,5]

box_plot_data=[Yes,No]
plt.figure(figsize=(12,8))
box = plt.boxplot(box_plot_data,patch_artist=True,flierprops=dict(color='#603371', markeredgecolor='#603371'),labels=['Would Reapply','Would Not Reapply'])

colors = ['#C067E2','#C067E2'] # Outlier colour = #603371, Median colour = k

for patch, color in zip(box['boxes'],colors):
    patch.set_facecolor(color)

for median in box['medians']:
    median.set(color='k', linewidth=3)
    
plt.title('Expectation Match vs. Reapplication Rate [n=90]')
plt.ylabel("How Closely 1A Experience Matched Expectations")

plt.show()
```
