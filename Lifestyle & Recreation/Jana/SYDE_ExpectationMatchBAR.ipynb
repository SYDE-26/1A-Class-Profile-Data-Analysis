```python
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['1', '2', '3', '4', '5', '6', '7']

# n = 90

No = [1,2,2,2,4,0,0] # Colour is #C067E2/#749BFF
Yes = [0,9,10,15,28,14,3] # Colour is #d294ea/#F7BF00

NoLabels = ['1%','2%','2%','2%','4%','0%','0%'] # Colour is #C067E2/#749BFF
YesLabels = ['0%','11%','11%','17%','31%','16%','3%'] # Colour is #d294ea/#F7BF00

X_axis = np.arange(len(X))
width = 0.4

fig, ax = plt.subplots(figsize=(14,8))
rects1 = plt.bar(X_axis - width/2, No, 0.4, color='#749BFF', label = 'Would Not Reapply')
rects2 = plt.bar(X_axis + width/2, Yes, 0.4, color='#F7BF00', label = 'Would Reapply')

  
plt.xticks(X_axis, X)
plt.xlabel("Expectation Match Rating")
plt.ylabel("Number of Responses")
plt.title("Program Expectation Match vs. Reapplication Rate [n=90]")
ax.bar_label(rects1, labels=NoLabels, padding=3)
ax.bar_label(rects2, labels=YesLabels, padding=3)
plt.legend()
plt.text(3.15, -4, "*NOTE: Percentages are based on TOTAL sample size (n=90) | Ratings: 1 = Completely Different, 7 = Exactly As Imagined", fontsize = 12, fontweight='semibold', ha='center')

plt.show()
```
