```python
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['Drunk Alcohol','Smoked/Vaped', 'Done Drugs', 'Had First Kiss','Had Sex']

No = [37,75,79,58,81] 
Yes = [54,16,12,33,10]

X_axis = np.arange(len(X))

plt.figure(figsize=(15,8))

plt.bar(X_axis - 0.2, No, 0.4, label = 'No')
plt.bar(X_axis + 0.2, Yes, 0.4, label = 'Yes')

  
plt.xticks(X_axis, X)
plt.xlabel("Event")
plt.ylabel("Number of Students")
plt.title("Number of Students Who Did X Event BEFORE 1A")
plt.legend()
plt.show()
```
