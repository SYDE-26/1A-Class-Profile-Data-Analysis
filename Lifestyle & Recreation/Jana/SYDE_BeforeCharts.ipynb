```python
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['Drunk Alcohol','Smoked/Vaped', 'Done Drugs', 'Had First Kiss','Had Sex']

No = [37,75,79,58,81] # Colour is #C067E2
Yes = [54,16,12,33,10] # Colour is #d294ea

NoLabels = ['41%','82%','87%','64%','89%']
YesLabels = ['59%','18%','13%','36%','11%']

X_axis = np.arange(len(X))
width = 0.4

fig, ax = plt.subplots(figsize=(14,8))
rects1 = plt.bar(X_axis - width/2, No, 0.4, color='#C067E2', label = 'No')
rects2 = plt.bar(X_axis + width/2, Yes, 0.4, color='#d294ea', label = 'Yes')

  
plt.xticks(X_axis, X)
plt.xlabel("Event")
plt.ylabel("Number of Students")
plt.title("Number of Students Who Did X Event BEFORE 1A [n=91]")
ax.bar_label(rects1, labels=NoLabels, padding=3)
ax.bar_label(rects2, labels=YesLabels, padding=3)
plt.legend()

plt.show()
```
