```python
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['Drunk Alcohol','Smoked/Vaped', 'Done Drugs', 'Had First Kiss','Had Sex', 'Attended a University Party']

No = [27,71,75,67,76,38] # Colour is #C067E2
YesBefore = [45,10,8,15,8,6] # Colour is #d294ea
YesFirst = [19,9,6,7,6,48] # Colour is #dfb3f0

NoLabels = ['30%','79%','83%','74%','84%','42%']
YesBeforeLabels = ['50%','11%','9%','17%','9%','7%']
YesFirstLabels = ['21%','10%','7%','8%','7%','53%']

X_axis = np.arange(len(X))
width = 0.4

fig, ax = plt.subplots(figsize=(14,8))
rects1 = plt.bar(X_axis - 0.2, No, 0.4, color='#C067E2', label = 'No')
rects2 = plt.bar(X_axis + 0.2, YesBefore, 0.4, color='#d294ea', label = 'Yes, Done It Before')
rects3 = plt.bar(X_axis + 0.2, YesFirst, 0.4, bottom=YesBefore, color='#dfb3f0', label = 'Yes, For the First Time')

  
plt.xticks(X_axis, X)
plt.xlabel("Event")
plt.ylabel("Number of Students")
plt.title("Number of Students Who Did X Event DURING 1A [n=90]")
ax.bar_label(rects1, labels=NoLabels, padding=3)
ax.bar_label(rects2, labels=YesBeforeLabels, padding=3)
ax.bar_label(rects3, labels=YesFirstLabels, padding=3)

plt.legend()
plt.show()
```
