```python
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['Drunk Alcohol','Smoked/Vaped', 'Done Drugs', 'Had First Kiss','Had Sex', 'Attended a University Party']

No = [27,71,75,67,76,38] 
YesBefore = [45,10,8,15,8,6]
YesFirst = [19,9,6,7,6,48]

X_axis = np.arange(len(X))

plt.figure(figsize=(15,8))

plt.bar(X_axis - 0.2, No, 0.4, label = 'No')
plt.bar(X_axis + 0.2, YesBefore, 0.4, label = 'Yes, Done It Before')
plt.bar(X_axis + 0.2, YesFirst, 0.4, bottom=YesBefore, label = 'Yes, For the First Time')

  
plt.xticks(X_axis, X)
plt.xlabel("Event")
plt.ylabel("Number of Students")
plt.title("Number of Students Who Did X Event DURING 1A")
plt.legend()
plt.show()
```
