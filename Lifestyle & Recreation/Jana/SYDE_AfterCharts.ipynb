```python
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['Drank Alcohol','Smoked/Vaped', 'Done Drugs', 'Had First Kiss','Had Sex', 'Attended a University Party']

No = [27,71,75,67,76,38] # Colour is #C067E2/#749BFF
YesBefore = [45,10,8,15,8,6] # Colour is #d294ea/#F7BF00
YesFirst = [19,9,6,7,6,48] # Colour is #dfb3f0/#B6AD80

NoLabels = ['30%','79%','83%','74%','84%','41%']
YesBeforeLabels = ['50%','11%','10%','17%','9%','7%']
YesFirstLabels = ['20%','10%','7%','9%','7%','52%']

X_axis = np.arange(len(X))
width = 0.4

fig, ax = plt.subplots(figsize=(14,8))
rects1 = plt.bar(X_axis - 0.2, No, 0.4, color='#749BFF', label = 'No')
rects2 = plt.bar(X_axis + 0.2, YesBefore, 0.4, color='#F7BF00', label = 'Yes, Done It Before')
rects3 = plt.bar(X_axis + 0.2, YesFirst, 0.4, bottom=YesBefore, color='#B6AD80', label = 'Yes, For the First Time')

  
plt.xticks(X_axis, X)
plt.xlabel("Event")
plt.ylabel("Number of Students")
plt.title("Number of Students Who Did X Event DURING 1A [n=90]")
ax.bar_label(rects1, labels=NoLabels, padding=3)
ax.bar_label(rects2, labels=YesBeforeLabels, padding=3)
ax.bar_label(rects3, labels=YesFirstLabels, padding=3)

plt.legend()
plt.show()
```
