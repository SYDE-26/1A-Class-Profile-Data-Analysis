```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()

size = 0.3
valsBig = np.array([[89, 89], [11, 11]])
valsSmall = np.array([[84.4,84.4], [8.89,8.89], [6.67,6.67]])

ax.pie(valsBig.sum(axis=1), radius=1,
       wedgeprops=dict(width=size, edgecolor='w'))


ax.pie(valsSmall.sum(axis=1), radius=1-size,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.set(aspect="equal", title='Sex Before and During 1A')

plt.show()
```
