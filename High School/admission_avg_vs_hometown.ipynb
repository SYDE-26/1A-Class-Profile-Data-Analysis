```python
# Import dependecies 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data

data = pd.read_csv('data - Sheet1 (5).csv')

# Identify columns of interest and rename them

column = ['avg', 'home']
relevant_data = data[column].rename(columns={'avg' : 'averages', 'home': 'hometowns'})
```

```python
# Plot graph with appropriate title, color palette, and data

sns.set(rc={'figure.figsize': (21,11)})
graph = sns.boxplot(
    data=relevant_data, x="hometowns", y="averages", color ="#C067E2" 
)
graph.set_title('Admission Averages vs. Hometowns [n=91]')
graph.set_xlabel('Region')
graph.set_ylabel('Top 6 Admission Average [including prerequisites]')
```
