```python
import matplotlib.pyplot as plt
import numpy as np
  
# create data
x = ['In Person Synchronous', 'Online Synchronous', 'Online Asynchronous']
y1 = np.array([35,25,25])
y2 = np.array([15,28,25])
y3 = np.array([15,20,23])
y4 = np.array([21,13,13])

y1Labels = ['41%','29%','29%']
y2Labels = ['17%','33%','29%']
y3Labels = ['17%','23%','27%']
y4Labels = ['25%','15%','15%']
  
# plot bars in stack manner
fig, ax = plt.subplots(figsize=(12,8))
rect1 = plt.bar(x, y1, 0.5, color='#FE8C5A')
rect2 = plt.bar(x, y2, 0.5, bottom=y1, color='#FF76A5')
rect3 = plt.bar(x, y3, 0.5, bottom=y1+y2, color='#FF8180')
rect4 = plt.bar(x, y4, 0.5, bottom=y1+y2+y3, color='#7F462D')
plt.xlabel("Lecture Format")
plt.ylabel("Number of Responses")

ax.bar_label(rect1, labels=y1Labels, padding=3)
ax.bar_label(rect2, labels=y2Labels, padding=3)
ax.bar_label(rect3, labels=y3Labels, padding=3)
ax.bar_label(rect4, labels=y4Labels, padding=3)

plt.legend(["0-25%","26-50%","51-75%","76-100%"])
plt.title("Percentage of Lectures Attended by Format [n=86]")
plt.show()
```
