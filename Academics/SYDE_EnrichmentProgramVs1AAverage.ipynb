```python
import matplotlib.pyplot as plt

No = [85,85,86.5,86.36,87,72,93,86.27,3.7,89,81,87,85,79,83,80,87,87.82,89,92.36,81,87.91,81,88,90,96,94.18,77,87,87,90.54]
AP = [91,72,89,89,83.09,87.91,81.6,86.18,85.46,92,84,86,87,78.42,85.64,89,87,87]
IB = [88,94,96,90,82,90,87,80.6,88.82,93,93,95,90,87,81.5,73]
Private = [83,90,83,86.82,82.7,78]

box_plot_data=[No, AP, IB, Private]
fig, ax = plt.subplots(figsize=(12,8))
box = plt.boxplot(box_plot_data,whis=0.5,patch_artist=True,flierprops=dict(color='#FF76A5', markeredgecolor='#9d0036'),labels=['None', 'AP', 'IB', 'Private School'])

colors = ['#FF76A5','#FF76A5', '#FF76A5', '#FF76A5', '#FF76A5'] # Outlier colour = #9d0036, Median colour = k
ax.set_ylim(ymin=70)

for patch, color in zip(box['boxes'],colors):
    patch.set_facecolor(color)

for median in box['medians']:
    median.set(color='k', linewidth=3)
    
plt.title('Highschool Enrichment Program vs. 1A Average [n=71]')
plt.ylabel("1A Average (%)")
plt.xlabel("Program")


plt.show()
```
