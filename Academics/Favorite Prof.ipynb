```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
from os import path
```

```python
prof_data = pd.read_csv("unordered.csv")
rel_cols = ['Favourite Professor?']
df = prof_data[rel_cols].dropna().rename(columns={'Favourite Professor?': 'prof'})
df
```

```python
grouped = df.groupby('prof')['prof'].count()
grouped
```

```python
text = " ".join(prof for prof in df.prof)
```

```python

ax = wordcloud = WordCloud(background_color="white",max_font_size=45, max_words= len(text),relative_scaling=.6,colormap='Reds_r').generate(text)
plt.figure()
plt.imshow(wordcloud,interpolation='bilinear')
plt.tight_layout(pad=0)
plt.axis("off")
plt.show()
```

```python
ax.to_file('graphs/FAV_prof.png')
```
