```python
#Household income vs Pay
import matplotlib.pyplot as plt
import numpy as np

# Random test data
data1 = [16.3, 19, 20]
data2 = [18.5, 18.5, 18.5, 18.5, 19, 19, 20, 20, 20.19, 21, 21, 21, 22, 22, 22, 23, 26.67, 27, 27, 28]
data3 = [16.67, 16.67, 17.25, 19, 19, 19, 21, 22, 22, 22.5, 23, 23.25, 23.5, 24.54, 26, 28.25]
data4 = [16.5, 17, 17.5, 18, 18, 18, 19, 20, 20, 20, 20.75, 22, 23, 25, 26, 27]
data5 = [18, 18.5, 18.5, 18.75, 20.5, 21, 23, 23, 24, 24, 25, 28, 28]
data6 = [17.25, 18.5, 20.19, 24, 28]
data7 = [16.43, 18, 19, 19, 20, 20, 20.19, 27.5]
all_data = [data1, data2, data3, data4, data5, data6, data7]
labels = ['0-50K', '50-100K', '100-150K', '150-200K', '200-250K', '250-300K', '300K+']

fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(9, 4))

# rectangular box plot
bplot1 = ax1.boxplot(all_data,
                     vert=True,  
                     showfliers = True, # vertical box alignment
                     patch_artist=True,  # fill with color
                     boxprops=dict(facecolor='#4C7DFA', color='#4C7DFA'),
                     
                     flierprops = dict(marker='o', markerfacecolor='#4C7DFA', markersize=7,
                  linestyle='none'),
                     labels=labels)  # will be used to label x-ticks
ax1.set_title('Household Income vs Hourly Coop Pay [81 respondents]')


# adding horizontal grid lines
for ax in [ax1]:
    #ax.yaxis.grid(True)
    ax.set_xlabel('Household Income (CAD)')
    ax.set_ylabel('Hourly Coop Pay (CAD/hr)')



plt.show()
ax1.figure.savefig('HouseholdIncomevsHourlyPay.svg')

```
