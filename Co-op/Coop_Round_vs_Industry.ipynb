```python
# Coop round vs Industry
import matplotlib.pyplot as plt
import numpy as np

labels = ['1st Round', '2nd Round', 'Continuous', 'Direct Offer']
automotiverobotics = np.array([1*(100/89), 0*(100/89), 2*(100/89),0*(100/89)])
bankingfinancial = np.array([10*(100/89),6*(100/89),3*(100/89),0*(100/89)])
education = np.array([0*(100/89),0*(100/89),4*(100/89),0*(100/89)])
government= np.array([1*(100/89),4*(100/89),3*(100/89),0*(100/89)])
health = np.array([7*(100/89),0*(100/89),1*(100/89),0*(100/89)])
manufacturing = np.array([0*(100/89),3*(100/89),3*(100/89),1*(100/89)])
marketing = np.array([1*(100/89),1*(100/89),2*(100/89),0*(100/89)])
retail = np.array([0*(100/89),2*(100/89),2*(100/89),0*(100/89)])
software = np.array([10*(100/89),4*(100/89),9*(100/89),2*(100/89)])
other = np.array([1*(100/89),3*(100/89),3*(100/89),0*(100/89)])

width = 0.5     

fig, ax = plt.subplots()


ax.bar(labels, automotiverobotics, width, bottom = bankingfinancial+education+government+health+manufacturing+marketing+retail+software+other, label='Automotive/Robotics', color ='#005660')
ax.bar(labels, bankingfinancial, width, bottom = education+government+health+manufacturing+marketing+retail+software+other, label='Banking/Financial', color = '#239389')
ax.bar(labels, education, width, bottom = government+health+manufacturing+marketing+retail+software+other, label='Education', color = '#57c3ad')
ax.bar(labels, government, width, bottom = health+manufacturing+marketing+retail+software+other, label='Government', color = '#7adfb5')
ax.bar(labels, health, width, bottom = manufacturing+marketing+retail+software+other, label='Health', color = '#08ebde')
ax.bar(labels, manufacturing, width, bottom = marketing+retail+software+other, label='Manufacturing', color = '#D6EAF8')
ax.bar(labels, marketing, width, bottom = retail+software+other, label='Marketing', color = '#85C1E9')
ax.bar(labels, retail, width, bottom = software+other, label='Retail', color = '#1E90FF')
ax.bar(labels, software, width, bottom = other, label='Software', color = '#0000ff')
ax.bar(labels, other, width, label='Other', color = '#002D82')

ax.set_xlabel('Coop Round')
ax.set_ylabel('Percentage of Respondents')
ax.set_title('Industry hiring patterns [89 respondents]')
ax.legend(bbox_to_anchor=(1.45, 1))


plt.show()
```
