```python
import numpy as np
import matplotlib.pyplot as plt

X = ['Software', 'PM', 'Data Science/ML', 'Product Design', 'UI/UX', 'Hardware', 'Information Security']
males = [36,3,3,1,1,0,1]
females = [26,6,1,3,7,1,1]
nonb = [1,0,0,0,0,0,0]

X_axis = np.arange(len(X))

plt.figure(figsize=(15,8))

plt.bar(X_axis - 0.1, males, 0.2, label = 'Males', color = '#4C7DFA')
plt.bar(X_axis + 0.1, females, 0.2, label = 'Females', color = '#5d4cfa')
plt.bar(X_axis + 0.3, nonb, 0.2, label = 'Non-Binary', color = '#373ded')

plt.xticks(X_axis, X)
plt.xlabel("Co-op Type")
plt.ylabel("Number of Students")
plt.title("Gender vs Type of Co-op [n=91]")
plt.legend()
plt.show()
```
