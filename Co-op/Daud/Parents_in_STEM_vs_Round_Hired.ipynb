```python
import numpy as np
import matplotlib.pyplot as plt

X = ['1st Round', '2nd Round', 'Continuous', 'Direct Offer']
Y_in_STEM = [21,13,18,2]
Y_not_in_STEM = [11,11,14,1]

X_axis = np.arange(len(X))

plt.figure(figsize=(15,8))

plt.bar(X_axis - 0.2, Y_in_STEM, 0.4, label = 'Parents in STEM', color = '#4C7DFA')
plt.bar(X_axis + 0.2, Y_not_in_STEM, 0.4, label = 'Parents Not in STEM', color = '#5d4cfa')

ax = plt.axes()
plt.xticks(X_axis, X)
ax.set_yticks([0,5,10,15,20])
plt.xlabel("Round Hired")
plt.ylabel("Number of Students")
plt.title("Parents in STEM vs Round Hired [n=91]")
plt.legend()
plt.show()
```
