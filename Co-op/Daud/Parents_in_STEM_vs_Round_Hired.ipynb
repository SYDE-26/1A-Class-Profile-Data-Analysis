```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['1st Round', '2nd Round', 'Continuous', 'Direct Offer']
in_STEM = [23.1,14.3,19.8,2.2]
not_in_STEM = [12.1,12.1,15.4,1.1]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars



fig, ax = plt.subplots(figsize=(15, 8))
rects1 = ax.bar(x - width/2, in_STEM, width, label='Parents in STEM', color = '#f7bf00')
rects2 = ax.bar(x + width/2, not_in_STEM, width, label='Parents Not in STEM', color = '#fe8c5a')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage of Students (%)')
ax.set_xlabel('Round Hired')
ax.set_title('Parents in STEM vs Round Hired [n=91]')
ax.set_xticks(x, labels)
ax.set_yticks([0,5,10,15,20])
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)


fig.tight_layout()

plt.show()
```
