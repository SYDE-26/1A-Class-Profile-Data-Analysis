```python
import numpy as np
import matplotlib.pyplot as plt

X = ['1st Round', '2nd Round', 'Continuous', 'Direct Offer']
males = [23.02, 20.77, 19.93, 20.15]
females = [21.74,21.24,22.48,18]
nonb = [0,0,18.00,0]

X_axis = np.arange(len(X))

plt.figure(figsize=(15,8))

plt.bar(X_axis - 0.1, males, 0.2, label = 'Males', color = '#4C7DFA')
plt.bar(X_axis + 0.1, females, 0.2, label = 'Females', color = '#5d4cfa')
plt.bar(X_axis + 0.3, nonb, 0.2, label = 'Non-Binary', color = '#373ded')

plt.xticks(X_axis, X)
plt.xlabel("Round Hired")
plt.ylabel("Average Pay ($CAD/hr)")
plt.title("Gender vs Pay vs Round [n=86]")
plt.legend()
plt.show()
```
