```python
# Coop round vs Type
import matplotlib.pyplot as plt
import numpy as np

labels = ['1st Round', '2nd Round', 'Continuous', 'Direct Offer']
productprojectmanagement = np.array([3*(100/90), 2*(100/90), 3*(100/90),1*(100/90)])
productdesign = np.array([1*(100/90),2*(100/90),1*(100/90),0*(100/90)])
uxui = np.array([3*(100/90),4*(100/90),1*(100/90),0*(100/90)])
software = np.array([23*(100/90),14*(100/90),23*(100/90),2*(100/90)])
data = np.array([0*(100/90),0*(100/90),2*(100/90),0*(100/90)])
hardware = np.array([0*(100/90),0*(100/90),1*(100/90),0*(100/90)])
informationsecurity = np.array([2*100/90, 0,0,0 ])

width = 0.25     # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, software, width, bottom = productprojectmanagement+productdesign+uxui+data+hardware+informationsecurity, label='Software', color = '#e32925')
ax.bar(labels, productprojectmanagement, width, bottom = productdesign+uxui+data+hardware+informationsecurity, label='Product/Project Management', color = '#f58132')
ax.bar(labels, productdesign, width, bottom = uxui+data+hardware+informationsecurity, label='Product Design', color = '#fee000')
ax.bar(labels, uxui, width, bottom = data+hardware+informationsecurity, label='UX/UI', color = '#4fb748')
ax.bar(labels, data, width, bottom = hardware+informationsecurity, label='Data Science & ML', color = '#6fc7b3')
ax.bar(labels, hardware, width, bottom = informationsecurity, label='Hardware', color = '#07446b')
ax.bar(labels, informationsecurity, width, label='Information Security', color = '#653F98')

ax.set_xlabel('Round Hired')
ax.set_ylabel('Percentage of Respondents (%)')
ax.set_title('Types of Jobs Accepted in 1A [90 respondents]')
ax.legend(bbox_to_anchor=(1.57, 1))

plt.show()
```
