```python
# Coop round vs Industry
import matplotlib.pyplot as plt
import numpy as np

labels = ['1st Round', '2nd Round', 'Continuous', 'Direct Offer']
automotiverobotics = np.array([1*(100/89), 0*(100/89), 2*(100/89),0*(100/89)])
bankingfinancial = np.array([10*(100/89),6*(100/89),3*(100/89),0*(100/89)])
education = np.array([0*(100/89),0*(100/89),4*(100/89),0*(100/89)])
government= np.array([1*(100/89),4*(100/89),3*(100/89),0*(100/89)])
health = np.array([7*(100/89),0*(100/89),1*(100/89),0*(100/89)])
manufacturing = np.array([0*(100/89),3*(100/89),3*(100/89),1*(100/89)])
marketing = np.array([1*(100/89),1*(100/89),2*(100/89),0*(100/89)])
retail = np.array([0*(100/89),2*(100/89),2*(100/89),0*(100/89)])
software = np.array([10*(100/89),4*(100/89),9*(100/89),2*(100/89)])
other = np.array([1*(100/89),3*(100/89),3*(100/89),0*(100/89)])

width = 0.5     

fig, ax = plt.subplots()


ax.bar(labels, automotiverobotics, width, bottom = bankingfinancial+education+government+health+manufacturing+marketing+retail+software+other, label='Automotive/Robotics', color ='#e32925')
ax.bar(labels, bankingfinancial, width, bottom = education+government+health+manufacturing+marketing+retail+software+other, label='Banking/Financial', color = '#f58132')
ax.bar(labels, education, width, bottom = government+health+manufacturing+marketing+retail+software+other, label='Education', color = '#fee000')
ax.bar(labels, government, width, bottom = health+manufacturing+marketing+retail+software+other, label='Government', color = '#88d531')
ax.bar(labels, health, width, bottom = manufacturing+marketing+retail+software+other, label='Health', color = '#4fb748')
ax.bar(labels, manufacturing, width, bottom = marketing+retail+software+other, label='Manufacturing', color = '#6fc7b3')
ax.bar(labels, marketing, width, bottom = retail+software+other, label='Marketing', color = '#07a7e0')
ax.bar(labels, retail, width, bottom = software+other, label='Retail', color = '#07446b')
ax.bar(labels, software, width, bottom = other, label='Software', color = '#653F98')
ax.bar(labels, other, width, label='Other', color = '#d23c95')

ax.set_xlabel('Round Hired')
ax.set_ylabel('Percentage of Respondents (%)')
ax.set_title('Industry hiring patterns [89 respondents]')
ax.legend(bbox_to_anchor=(1.45, 1))


plt.show()
```
